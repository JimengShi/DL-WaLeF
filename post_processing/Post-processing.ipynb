## MAE vs. prediction length

```python
def mae(y_true, predictions):
    y_true, predictions = np.array(y_true), np.array(predictions)
    return np.mean(np.abs(y_true - predictions))
```

```python
print(mae(true_to_seires, mlp))
print(mae(true_to_seires, rnn))
print(mae(true_to_seires, lstm))
print(mae(true_to_seires, conv_mlp))
print(mae(true_to_seires, conv_rnn))
```

## RMSE vs. prediction length

```python
def rmse(predictions, targets):
    targets, predictions = np.array(targets), np.array(predictions)
    return np.sqrt(np.mean((predictions-targets)**2))
```

```python
print(rmse(true_to_seires, mlp))
print(rmse(true_to_seires, rnn))
print(rmse(true_to_seires, lstm))
print(rmse(true_to_seires, conv_mlp))
print(rmse(true_to_seires, conv_rnn))
```
