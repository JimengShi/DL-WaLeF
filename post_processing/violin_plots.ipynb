```python
def concat_df(df, i):
    a = df.iloc[:, i:i+1].values
    b = df.iloc[:, i+1:i+2].values
    c = df.iloc[:, i+2:i+3].values
    d = df.iloc[:, i+3:i+4].values
    e = np.concatenate((a, b, c, d), axis=0)
    dataframe = pd.DataFrame(e)
    return dataframe
```
