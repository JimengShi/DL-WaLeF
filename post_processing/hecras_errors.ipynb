```python
import pandas as pd
import numpy as np

hecras = pd.read_csv('../data/ras_1920.csv')
hecras
```

```python
ras = hecras[['T2 RAS S1','T2 RAS S25A', 'T2 RAS S25B', 'RAS S26']]
obs = hecras[['S1 OBS', 'S25A_TW OBS' , 'S25B_TW OBS', 'S26_TW OBS']]
```

```python
from sklearn.metrics import mean_squared_error as mse
from sklearn.metrics import mean_absolute_error as mae
```

```python
print('MAE = {}'.format(float("{:.4f}".format(mae(ras, obs)))))
print('RMSE = {}'.format(float("{:.4f}".format(np.sqrt(mse(ras, obs))))))
```
